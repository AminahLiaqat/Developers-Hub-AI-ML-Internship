```python
import pandas as pd  # Pandas for loading and manipulating datasets
import seaborn as sns #Seaborn for creating statistical data visualizations
import matplotlib.pyplot as plt # Matplotlib for plotting graphs and charts
```

```python
iris = sns.load_dataset('iris') # Load the built-in Iris dataset from the Seaborn library
print(iris.head()) # Print first 5 rows
```

```python
print("Shape of datasets:") # Display the dimensions of the dataset (rows and columns)
print(iris.shape)
```

```python
print("Column Names:") # For coloumns names in the dataset 
print(iris.columns)
```

```python
print("First ten rows:") 
print(iris.head(10)) # To display first 10 rows in the dataset
```

```python
print("Dataset information:")
print(iris.info()) # To display information about the dataset, such as column names, data types, non-null values, and memory usage
```

```python
print("Dataset summary:")
print(iris.describe()) # Display summary statistics of the numerical features
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot( # Create a scatter plot to visualize the relationship between sepal length and sepal width for different iris species
    x='sepal_length',
    y='sepal_width',
    hue='species',
    data=iris
)

plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
iris.hist(figsize=(10,8)) ## Create histograms for all numerical features in the Iris dataset the figure size is set to 10 inches wide and 8 inches high
plt.suptitle('Histogram')
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(data=iris.iloc[:, :-1]) # Create box plots for all numerical features in the Iris dataset iris.iloc[:,:-1] selects all feature columns except the last column (species)

plt.title('Box Plot of Iris Features')
plt.show()
```
